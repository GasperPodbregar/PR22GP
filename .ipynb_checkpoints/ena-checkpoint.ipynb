```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pyaxis import pyaxis
import warnings
warnings.filterwarnings("ignore")

EXAMPLE_URL = 'https://pxweb.stat.si/SiStatData/Resources/PX/Databases/Data/05C4002S.PX'

px = pyaxis.parse(EXAMPLE_URL, encoding='ISO-8859-2')
print(px['DATA'].info())
```

```python
data = px['DATA']
data['DATA'] = pd.to_numeric(data['DATA'])
data.loc[data['SPOL'] == 'enske', 'SPOL'] = 'zenske'
data.loc[data['SPOL'] == 'Moki', 'SPOL'] = 'moski'

steviloM = data[(data['STAROST'] == 'Starost - SKUPAJ') & (data['OBČINE'] == 'SLOVENIJA') & (data['SPOL'] == 'moski')] 
steviloF = data[(data['STAROST'] == 'Starost - SKUPAJ') & (data['OBČINE'] == 'SLOVENIJA') & (data['SPOL'] == 'zenske')] 

stevilo_prebivalcev = data[(data['STAROST'] == 'Starost - SKUPAJ') & (data['OBČINE'] == 'SLOVENIJA') & (data['SPOL'] == 'Spol - SKUPAJ')] 
data = data[data['STAROST'] != 'Starost - SKUPAJ']

data['STAROST'] = data['STAROST'].str.removesuffix(" let")
data['STAROST'] = data['STAROST'].str.removesuffix(" leto")
data['STAROST'] = data['STAROST'].str.removesuffix(" leti")
data['STAROST'] = data['STAROST'].str.removesuffix(" leta")
data['STAROST'] = data['STAROST'].str.removesuffix(" lete")
data['STAROST'] = data['STAROST'].str.removesuffix(" +")
data['STAROST'] = pd.to_numeric(data['STAROST'])


slovenijaF = data[(data['OBČINE'] == 'SLOVENIJA') & (data['SPOL'] == 'zenske')]
slovenijaM = data[(data['OBČINE'] == 'SLOVENIJA') & (data['SPOL'] == 'moski')]
slovenijaS = data[(data['OBČINE'] == 'SLOVENIJA') & (data['SPOL'] == 'Spol - SKUPAJ')]

trbovlje = data[(data['OBČINE'] == 'Trbovlje') & (data['SPOL'] == 'Spol - SKUPAJ')]
```

```python
slovenijaS.boxplot(column="DATA", by="STAROST", figsize=(20,10))
```

```python
ymin, ymax = [0, 40000]
ax = slovenijaM.boxplot(column="DATA", by="STAROST", figsize=(20,10))
bx = slovenijaF.boxplot(column="DATA", by="STAROST", figsize=(20,10))
ax.set_ylim([ymin, ymax])
bx.set_ylim([ymin, ymax])
```

```python
ax = plt.gca()

stevilo_prebivalcev.plot(kind='line',
        x='POLLETJE',
        y='DATA',
        color='green', ax=ax)
plt.title('LinePlots')

plt.show()
```

```python
rast = dict()
prvi = stevilo_prebivalcev.iloc[0, stevilo_prebivalcev.columns.get_loc('DATA')]
prev = prvi
for x, y in zip(stevilo_prebivalcev['DATA'], stevilo_prebivalcev['POLLETJE']):
    rast[y] = (x/prev - 1) * 100
    prev = x

skupna_povprecna_rast = (prev/prvi - 1) * 100
print(skupna_povprecna_rast)
r = pd.Series(rast, name='DATA')
r.index.name = 'POLLETJE'
r.reset_index()
r.plot.bar(x='POLLETJE', y='DATA', figsize=(20,10))
```
