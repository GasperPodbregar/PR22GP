## Predstavitev podatkov

Zbirka podatkov se nahaj na naslednji spletni strai https://podatki.gov.si/dataset/surs05c4002s

Množica podatkov vsebuje število prebivalcev po občinah v Sloveniji, razdeljeno po starostnih skupinah. Podatki si sledijo po polletjih (2008H1, 2008H2,...). 

 
| Stolpci       | Opis          |
| ------------- |:-------------:|
| Spol          | Spol (skupaj, m, ž) |
| Občine        | Podatki po občinah      |
| Polletje      | Podatki po polletju (2008 - 2020)     |
| Starost       | Starostne skupine              |
| Data          | Število prebivalcev v skupini |

Podatki so v obliki PX, vsaka celica pa predstavlja število ljudi ki ustreza pogojem. Datoteko sem najprej odprl in prevedel podatke v obliko DataFrame (knjižnjica pandas). Ta DataFrame sem naknadno razbil na več različnih tabel in slovarjev.

Prvoten cilj je bil določiti najstarejše občine (po povprečni starosti) in najmlajše, rast prebivalstva po občinah in vpliv drugih faktorjev na število prebivalstva. Težave sem imel glede šumnikov ter manjkajočih podatkov za 2 občini do leta 2012.

## 1. Slovenija

Spodnji graf prikaže povprečno število prebivalcev po starostnih skupinah od leta 2008 do 2020.




Starostna struktura Slovenije predstavlja stagnacijo prebivalstva. Izjema je starostna skupina 85+ ki predstavlja izjemno velik delež prebivalstva. Pričakovana življenska doba v Sloveniji je za moške ~ 78 let, ženske pa ~ 84 (2020). Ta podatek se zelo pozna če graf razdelimo na ženske in moške posebej.



V povprečju je zadnjih 15 let v Sloveniji 3x več žensk v starostni skupini 85+ kot moških.

## 4. Rast / upad

Slovenija že kar nekaj let stagnira glede prebivalstva vendar so leta kjer rast / upad izstopa zaradi zunanjih dejavnikov. Povprečna rast 2008 do 2020 je 3.65%





Na rast prebivalstva vpliva povprečna starost prebivalstva. Spodaj je pikazan graf kjer se lahko razvidi neko razmerje vendar ni tako izrazito zaradi priseljevanja migrantov. X os predstavlja povprečno starost občine, y os pa povprečno rast prebivalstva.

## Zaključek

Slovenija ima normalno starostno strukturo in število prebivalstva stagnira. Podatki najbolj izstopajo leta 2008 in 2009 zaradi začetka globalne krize (the Great Recession). Največja rast se je zgodila 2019, kasneje 2020 pa se je zgodila pandemija.
